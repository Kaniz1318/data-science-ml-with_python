**4. What are the max and min ages of your customer? Can you find mean of your customer?**

```python
# Code here please
max_age = df['age'].max()
min_age = df['age'].min()
avg_age = df['age'].mean()

print("Max. age of the customer is: ", max_age)
print("Min. age of the customer is: ", min_age)
print("Avg. age of the customer is: ", avg_age)
```

**5. What are the three most common customer's names?**<br>
&#9989; [<code>**value_counts()**</code>](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) returns object containing counts of unique values. The resulting object will be in descending order so that the first element is the most frequently-occurring element. Excludes NA values by default.

```python
# Code here please
df['first'].value_counts().head(3)
```

**6. Two customers have the same phone number, can you find those customers**?

```python
# Code here please
# A) Phone numbers that appear exactly twice
dup_twice = df['phone'].value_counts()
dup_twice[dup_twice.eq(2)]
# B) The "top 2" phone numbers by frequency
df['phone'].value_counts().head(2)

```

```python
# lets find what phone number is twice!
```

```python
# Code here please
# Find the duplicate phone and show all details for those customers
dup_phone = df['phone'].value_counts().loc[lambda s: s.eq(2)].index[0]
df[df['phone'].eq(dup_phone)][
    ['prefix','first','last','email','gender','age','company','profession','phone',
     'postal','province','cc_no','cc_exp','cc_type','price(CAD)','fav_color','ip','weekday','ampm','date']
].sort_index()
```

```python
# Now we know the phone number, let's find out the other stuff!
```

**7. How many customers have profession "Structural Engineer"?**

```python
# Code here please
mask = df['profession'].eq('Structural Engineer')
df[mask].count()
```

**8. How many male customers are 'Structural Engineer'?**

```python
# Code here please
df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Male')].count()
```

**9. Find out the female Structural Engineers from province Alberta (AB)?**

```python
# Code here please
res = df[
    (df['profession'] == 'Structural Engineer') &
    (df['gender'] == 'Female') &
    (df['province'] == 'AB')
]
res
```

**10. What is the max, min and average spending?**

```python
# Code here please
mx = df['price(CAD)'].max()
mn = df['price(CAD)'].min()
avg = df['price(CAD)'].mean()

print("Max. spending: ", mx)
print("Min. spending: ", mn)
print("Avg. spending: ", f"{avg:.15f}")
```

**11. Who did not spend anything? Company wants to send a deal to encourage the customer to buy stuff!**

```python
# Code here please
zero_spend = df[df['price(CAD)'].eq(0.0)]
zero_spend[
    ['prefix','first','last','email','gender','age','company','profession','phone',
     'postal','province','cc_no','cc_exp','cc_type','price(CAD)','fav_color','ip','weekday','ampm','date']
]
```

**12. As a loyalty reward, company wants to send thanks coupon to those who spent 100CAD or more, please find out the customers?**

```python
# Code here please
high_spenders = df[df['price(CAD)'] >= 100.0]
high_spenders[
    ['prefix','first','last','email','gender','age','company','profession','phone',
     'postal','province','cc_no','cc_exp','cc_type','price(CAD)','fav_color','ip','weekday','ampm','date']
]
```

**13. How many emails are associated with this credit card number '5020000000000230'?**

```python
# Code here please
mask = df['cc_no'].astype(str).eq('5020000000000230')
emails = df.loc[mask, 'email'].reset_index(drop=True)
print(emails)
```

**15. How many people use Visa as their Credit Card Provider?**

```python
# Code here please
visa_count = (df['cc_type'].str.strip().str.casefold() == 'visa').sum()
print(visa_count)
```

**16. Can you find the customer who spent 100 CAD using Visa?**

```python
# Code here please
res = df[
    (df['price(CAD)'] == 100.0) &
    (df['cc_type'].str.strip().str.casefold() == 'visa')
]
res[
    ['prefix','first','last','email','gender','age','company','profession','phone',
     'postal','province','cc_no','cc_exp','cc_type','price(CAD)','fav_color','ip','weekday','ampm','date']
]
```

**18. Can you tell the top 5 most popular email providers? (e.g. gmail.com, yahoo.com, etc...)**

```python
# Code here please
df['email'].str.split('@').str[-1].str.lower().value_counts().head(5)
```

**19. Is there any customer who is using email with "am.edu"?**<br>
Hint: Use `lambda` expression in `apply()`. split the email address at `@`.  

```python
# Code here please
has_am_edu = df['email'].apply(lambda e: e.split('@')[-1].lower() == 'am.edu')
df[has_am_edu][
    ['prefix','first','last','email','gender','age','company','profession','phone',
     'postal','province','cc_no','cc_exp','cc_type','price(CAD)','fav_color','ip','weekday','ampm','date']
]
```
